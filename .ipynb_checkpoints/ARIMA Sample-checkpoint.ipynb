```python
import itertools
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import warnings

from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv(r'data\NASDAQ-1-YEAR- DAILY.csv', index_col = 0)
df.head()

```

```python
data = df[['Adj Close']]
train_data = data[:200]
print(len(data))
test = data[200:]
```

```python
df_diff1 = data.diff().fillna(data)
data.plot()
df_diff2 = df_diff1.fillna(df_diff1)
df_diff2.plot()
```

```python
warnings.filterwarnings("ignore")
mod = sm.tsa.statespace.SARIMAX(train_data,
                                order=(0, 1, 0),
                                enforce_stationarity=False,
                                enforce_invertibility=False)

results = mod.fit()
```

```python
results.plot_diagnostics(figsize=(20, 14))
```

```python
pred0 = results.get_prediction(start='2021-04-13', dynamic=False)
pred0_ci = pred0.conf_int()
pred1 = results.get_prediction(start='2021-04-13', dynamic=True)
pred1_ci = pred1.conf_int()
```

```python
forecast = results.forecast()
```

```python
forecast
```

```python
pred2 = results.get_forecast(steps=50)
pred2_ci = pred2.conf_int()
```

```python
pred2_ci
```

```python
ax = data.plot(figsize=(20, 16))
pred0.predicted_mean.plot(ax=ax, label='1-step-ahead Forecast (get_predictions, dynamic=False)')
pred1.predicted_mean.plot(ax=ax, label='Dynamic Forecast (get_predictions, dynamic=True)')
pred2.predicted_mean.plot(ax=ax, label='Dynamic Forecast (get_forecast)')
ax.fill_between(pred2_ci.index, pred2_ci.iloc[:, 0], pred2_ci.iloc[:, 1], color='k', alpha=.1)
plt.ylabel('Adj Close')
plt.xlabel('Date')
plt.legend()
plt.show()
```

```python
predicted1, resid_test = [], []
history = train_data

for t in range(len(test)):
    model = ARIMA(history, order=(0,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    # !!!nable to access dataframe through index
    print(t, test.loc[[t]])
    resid_test.append(test[t] - output[0])
    predicted1.append(yhat)
    obs = test[t]
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
test_resid = []
for i in resid_test:
    test_resid.append(i[0])
error = mean_squared_error(test, predicted1)
print('Test MSE: %.3f' % error)
plt.plot(test)
plt.plot(predicted1, color='red')
plt.show()
```
